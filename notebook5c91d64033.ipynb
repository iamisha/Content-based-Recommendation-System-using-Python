**Content-Based Recommendation System using Python**

**TF*IDF Algorithm**

> Term Frequency (TF): For each document, the TF is calculated for each term. TF measures the frequency of a term within a document and is usually calculated as the ratio of the number of occurrences of a term to the total number of terms in the document. It helps identify the most important terms in a document.


> Inverse Document Frequency (IDF): IDF measures the rarity or importance of a term across the entire document collection. IDF is calculated as the logarithm of the ratio between the total number of documents in the collection and the number of documents that contain the term. Terms that appear in many documents have a lower IDF value, indicating that they are less important.


**Vector Space Model**

> The vector space model in Python represents documents as numerical vectors in a high-dimensional space. It involves preprocessing text, creating a vocabulary, calculating term frequencies (TF) and inverse document frequencies (IDF), and computing TF-IDF values. The resulting TF-IDF matrix enables document comparison and similarity calculations using techniques like cosine similarity. Python libraries like scikit-learn can assist in implementing the vector space model efficiently.


**Calculation of Cosine Similarities**
> Cosine similarity is a measure used to calculate the similarity between two vectors in a vector space. In the context of text document analysis using the vector space model, cosine similarity is often used to compare the similarity between two documents represented as TF-IDF vectors

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
```

```python
df_movies = pd.read_csv("/kaggle/input/tmdb-movie-metadata/tmdb_5000_movies.csv")
df_credits = pd.read_csv("/kaggle/input/tmdb-movie-metadata/tmdb_5000_credits.csv")
```

```python
# removing the english words like but, the, for, etc
tfidf = TfidfVectorizer(stop_words = "english")
# filling the removed english words with empty string
df_movies['overview'] = df_movies['overview'].fillna("")
# build the vector space model
tfidf_matrix = tfidf.fit_transform(df_movies['overview'])
```

```python
# develop the cosine similarities
cosine_similarities = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python

indices = pd.Series(df_movies.index, index=df_movies['original_title']).drop_duplicates()
indices
```

```python
indices["Pirates of the Caribbean: At World's End"]
```

```python
def get_recommendations(title, cosine_similarities = cosine_similarities ):
    idx = indices[title]
    sim_scores = enumerate(cosine_similarities[idx])
    # to sort based on the particular number
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]
    
    sim_index = [i[0] for i in sim_scores]
    print(df_movies["original_title"].iloc[sim_index])

```

```python
get_recommendations('The Dark Knight')
```
